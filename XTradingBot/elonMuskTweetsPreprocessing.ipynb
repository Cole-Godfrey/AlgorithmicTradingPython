```python
import pandas as pd
```

```python
# import csv to pandas dataframe
# noinspection PyArgumentList
df = pd.read_csv("elon_musk_tweets.csv", encoding="latin1")
```

```python
# select time and tweet columns
df = df[["date", "text"]]
```

```python
# Set order from least to most recent
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values(by="date").reset_index(drop=True)
```

```python
# removes all URLs (replaces with {URL} placeholder)
df["text"] = df["text"].str.replace(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '{URL}', regex=True)
```

```python
# change column names
df.columns = ["Time", "Tweet"]
```

```python
# export dataframe to csv
df.to_csv("muskTweetsPreProcessed.csv", index=False)
```
